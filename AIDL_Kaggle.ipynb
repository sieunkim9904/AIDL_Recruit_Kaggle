```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
```

```python
# Load data set
train_df=pd.read_csv('train.csv')
test_df=pd.read_csv('test.csv')

# Show first five rows of train.csv
train_df.head()
```

```python
# Check for bad/null data
train_df.isnull().sum()
```

```python
# Create the X and y arrays
X = train_df[["Watch time(Minutes)", "Stream time(minutes)", "Peak viewers", "Average viewers", "Followers", "Views gained", "Partnered", "Mature"]]
Y = train_df[["Followers gained"]]

print(X.shape)
print(Y.shape)
```

```python
# Create a scaler for the inputs and outputs
X_scaler = MinMaxScaler(feature_range=(0, 1))
Y_scaler = MinMaxScaler(feature_range=(0, 1))

# Scale the input and output data
X[X.columns] = X_scaler.fit_transform(X[X.columns])
Y[Y.columns] = Y_scaler.fit_transform(Y[Y.columns])

# Convert the inputs and outputs to numpy format
X = X.to_numpy()
Y = Y.to_numpy()
```

```python
# Define model
model = Sequential()
model.add(Dense(50, input_dim=8, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
# Compile model
model.compile(
    loss='mean_squared_error',
    optimizer='SGD'
)
```

```python
# Train model
model.fit(
    X,
    Y,
    epochs=50,
    batch_size=8,
    shuffle=True,
    verbose=2
)
```

```python
print("Model training results:")

# Report RMSE of the training
# Scale back the values
predictions_train = model.predict(X, verbose=0)

mse_train = mean_absolute_error(
    Y_scaler.inverse_transform(predictions_train),
    Y_scaler.inverse_transform(Y)
)
print(f" - Training Set Error: {mse_train}")
```

```python
# Now applying the method to test data
test_df.head()
```

```python
# Check for bad/null data
test_df.isnull().sum()
```

```python
# Create the X array for test data
X_test = test_df[["Watch time(Minutes)", "Stream time(minutes)", "Peak viewers", "Average viewers", "Followers", "Views gained", "Partnered", "Mature"]]
print(X_test.shape)
```

```python
# Scale the X array for the model
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Make prediction
predict_Y = model.predict(X_test_scaled)
```

```python
# Unscale the predicted Y
unscaled_predict_Y = Y_scaler.inverse_transform(predict_Y)

# unscaled_predict_Y is right now a 2d array of size[200][1]. 
# In order to convert the data into a DataFrame it should be a 1d array
# So turn 2d array of [200][1] to 1d array of [200]
unscaled_predict_Y_1d = unscaled_predict_Y.flatten()

#convert into csv
df = {'Channel': test_df['Channel'], 'followers gained': unscaled_predict_Y_1d}
result = pd.DataFrame(df)
print(result)
result.to_csv('submission_Sieun.csv',index=False)
```
